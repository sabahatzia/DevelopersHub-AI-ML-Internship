```python
import os
import sys
from google import genai
from google.genai import errors

def start_chatbot():
    # ==========================================
    # 1. PASTE YOUR API KEY HERE
    # ==========================================
    API_KEY = "AQ.Ab8RN6KrAwylAHG0a03kfZ5Xo5hm0ALOi3sSNBNlhgXOSq5T5w"
    
    if API_KEY == "YOUR_GEMINI_API_KEY_HERE" or not API_KEY:
        print("Error: Please replace 'YOUR_GEMINI_API_KEY_HERE' with your actual Gemini API key.")
        return

    # 2. Initialize the client directly with your key
    try:
        client = genai.Client(api_key=API_KEY)
    except Exception as e:
        print(f"Error initializing client. Details: {e}")
        return

    # 3. Start a continuous chat session
    # gemini-2.5-flash is fast, smart, and perfect for conversational chat.
    model_id = "gemini-2.5-flash"
    chat = client.chats.create(model=model_id)

    print("====================================================")
    print(f"Chatbot initialized using {model_id}!")
    print("Type your message in the 'You:' box and press Enter.")
    print("Type 'quit' to exit the chat cleanly.")
    print("====================================================\n")

    while True:
        try:
            # Get input from the user inside Jupyter
            user_message = input("You: ")
            
            # Check for exit command
            if user_message.strip().lower() == 'quit':
                print("\nGoodbye! Chat session closed.")
                break
            
            # Skip empty messages
            if not user_message.strip():
                continue

            # Send the message to the model and get the response
            response = chat.send_message(user_message)
            
            print(f"\nGemini: {response.text}\n")
            print("-" * 50)

        except errors.APIError as e:
            print(f"\nGemini API Error: {e}\n")
            break
        except KeyboardInterrupt:
            print("\nSession interrupted. Goodbye!")
            break

if __name__ == "__main__":
    start_chatbot()
```
